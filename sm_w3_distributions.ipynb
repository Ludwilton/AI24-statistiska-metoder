# discrete probability densities

täthetsfunktion
sannolikheten att vår stokastiska variab

$f(x) = P[X=x]$

necessary and sufficient conditions for a discrete density

$f(x) \leq 0$ (probability function. sv. täthetsfunktion)

$\sum f(x) = 1$



cumulative distribution function

$F(x) = P[X\leq x]$

### expected value (väntevärde):

$E[X]= \sum\limits_{x}xf(x)$

ex $1*\frac{1}{6}+2*\frac{1}{6}+3*\frac{1}{6}+4*\frac{1}{6}... = 3.5$

variance

Var X = $\sigma^2=[E(X-\mu)^2$

$\sigma^2=Var X = E[X^2] - (E[X])^2$

Standard deviation:

$\sigma = \sqrt{VarX}= \sqrt{\sigma^2}$

## Geometric distribution
en geometrisk distribution minskar och är ganska brant

X = the number of tries until first success
S = {S,FS,FFS,FFFS,...,$\infty S$}

$f(x) = (1-p)^{x-1}P$

$f(x) = 1-q^{[x]}$
 
$q = 1-p $

$[x] = \texttt{largest integer smaller than x}$

väntevärdet 
$E[x]=\frac{1}{p}$

Variansen

$Var X = \frac{q}{p^2}$


## binomial distribution
visuellt en bredare bell-curve

A fixed number of trials (n)

X = how many successes over the n trials

$F(x) = \binom{n}{x} p^x(1-p)^{n-x}$

$f(x)=\sum\limits^{{t}}_{x=0}\binom{n}{x}p^x(1-p)^{n-x}$

$E[x]=\mu=np$

$Var X = \sigma^2 = npq$

$q = 1-p$

## negative binomial distribution

X = number of trials until r successes

$$f(x)=\binom{x-1}{r-1}(1-p)^{x-r}p^r$$

x = r, r+1, r+3,...

väntevärde
$E[x]=\frac{r}{p}$

varians $Var(x)= \frac{rq}{p^2}$

$q=1-p$



## hypergeometric distribution
sannolikhetsfunktion 

användbar om man försöker veta vilken andel av ett stickprov som har en viss egenskap, (den måste dyka upp i populationen)


$$f(\alpha)=\frac{\binom{r}{x}\binom{n-r}{n-x}}{\binom{N}{n}}$$

N -  Number of objects

n - Sample size

r - number of objects with a particular trait

$E[x]=n(\frac{r}{N})$

varians 

$var X = n(\frac{r}{N})(\frac{N-r}{N})(\frac{N-n}{n-1})$

## Poisson distribution
används som en regressionsmetod

försöker hitta signaler under ett visst tidsintervall

täthetsfunktion

$$f(x)= \frac{e^{-k} k^x}{x!}$$

$k>0$, $x = 1,2,3,...$

väntevärde

$E[x]=k$

varians

$Var X = k$


$k=\lambda s$

$\lambda = \texttt{average occurences per time unit}$



## normal distrubition (continous distribution)
def:

a random variable X is continous if it can assume any real value in some interval $and$ the probability that it assumes any specific 0

$f (x) = P[X=x] \implies$ 0 everywhere?

1. $f(x)\geq 0$
2. $\int_x f(x)dx=1$
3. $p[asX\geq b]= \int_{\alpha}f(x)dx$


(standardisering?)

$$f(x)=\frac{1}{\sqrt{2\pi\sigma}}^{e^{-\frac{1}{2}((X-\mu)/\sigma)^2}}$$

$\mu, \sigma \implies \texttt{mean, standard deviation}$





### normal probability rule (68-95-99)

1 $\sigma$ = 68%

$P[-\sigma<X-\mu<\sigma] = .68$

2 $\sigma$ = 95%

$P[-2\sigma<X-\mu<2\sigma] = .95$

3 $\sigma$ = 99.7%

$P[-3\sigma<X-\mu<3\sigma] = .997$
