## Gemensamma distributioner (eng. joint distributions)

1. 
$ Fxy(X,Y) => 0 $ 

2.
$ \sum\limits_{x} \sum\limits_{y} Fxy(X,Y) = 1 $ 

#### marginell täthet

för x

$ Fx(x) = \sum\limits_{y}Fxy(X,Y) $

för Y

$ Fx(Y) = \sum\limits_{x}Fxy(X,Y) $

#### betingad täthet

$ fx|y = \frac{Fxy(X,Y)}{Fy(Y)} $ 

$ fy|x = \frac{Fxy(X,Y)}{Fx(X)} $
