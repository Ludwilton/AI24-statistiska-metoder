# Confidence intervals



$\alpha$ representerar sannolikheten att *ngot* faller inom intervallet

$100(1-\alpha)%$ interval

e.g

$\alpha = 0.05$

$1-\alpha = 0.95$

if we standardize and normalize data we get a standard normal distribution:



then a confidence interval is given by $\bar{X}\pm Z_{\alpha/2}(\alpha/\sqrt{n})$

in scipy $Z_{\alpha/2}$ is found with "ppf"


when mean($\mu$) and variance($\sigma^2$) is not known we use $\hat{X}\pm t_{\alpha/2}(s/\sqrt{n})$

note that $s/\sqrt{n}$ is not an unbiased estimate of $\sigma/\sqrt{n}$

for linear regression:

$\hat\beta_{i}\pm t_{\alpha/2}(S\sqrt{C_{ii}})$

e.g for a chosen $\alpha$
$b_{1} = 21.25 \pm 1.96$

# gemensamma frdelningar / Joint distributions


$Y=  0 \beta_{0}+\beta_{1}X_{1}+...+\beta_{d}X_{d}$

E, vntevrdet

$\sum E[X_{i}] = E[Y]$ mste vara en linjrkombination

we would like to show that E is a linear transformation(linjr avbildning)

$\implies$ $E[\mu X+vY]=\mu E[X]+vE[Y]$



vntevrdet, expected value for joint distributions

$E[x]=\sum\limits_{x}\sum\limits_{y}xf(x,y)$
$E[x]=\sum\limits_{x}\sum\limits_{y}yf(x,y)$

note:

$E[cX]=cE[x]$


$E[\mu X + vY] = \sum\limits_{x}\sum\limits_{y}(\mu x + vy)f(x, y)$

=

$\sum\limits_{x}\sum\limits_{y}\mu xf(x,y)+\sum\limits_{x}\sum\limits_{y}vyf(x,y)$

=

$E[\mu x] + E[vY]=\mu E[x]+vE[y]$


bevis att vntevrdet r linjrt ^^

Slumpvariabel X = utfallet av kast med en sex-sidig trning

ett stickprov med storlek 6 r [2,2,3,4,4,5]

detta blir en uniform distribution(alla har lika probabilitet)



formel 
$f(x)=\frac{n(A)}{n(S)}$

uniform frdelning p[X=x]= 1/6

*men* fr stickprovet:$(F(x)=\frac{f}{n})$

$P[X=1] = 0$

$P[X=2] = \frac{2}{6}=\frac{1}{3}$

X = antalet trningskast till vi slr en 6:a
ett frsk = [[2,4,4,3,5,6], [1,6], [2,5,4,6]]

sannolikhet att sl en sexa: P = 1/6

detta blir d en geometrisk frdelning med p = 1/6

funktionen ser ut:

$f(x) = (1-p)^{x-1}p$

X = antalet radarsignaler som identifieras under 30 minuter dr 10 signaler mottas
P = 1/2, sannolikheten att en signal korrekt identifieras

hitta sannolikheten att som mest 7 signaler identifieras korrekt:

detta blir en binomial frdelning

$X \approx Binom(n=10,p=,5$

$ P[X\leq 7]$ $\implies$ cdf(7, n=10, p=.5) 

```python
import scipy.stats as stats
stats.binom.cdf(7, n=10, p=.5),stats.binom.pmf(7, n=10, p=.5)
```

```python
# hypergeometrisk distribution
# 
# vid ltillverkning tillverkas 1000 flaskor per timma
# 
# varje timma tas ett stickprov med storlek 20 frn tillverkningen, och innehllets volym mts.
# 
# X = antalet underfyllda flaskor
# 
# antag att vid en viss timma produceraas 100 underfyllda flaskor.
# 
# vad r sannolikheten att $\geq 3$ underfylld flaska frekommer i stickprovet
# 
# $ P[X\geq 3] \implies cdf$ 

```
