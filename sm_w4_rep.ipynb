```python
df = pd.read_csv("./data/Auto.csv")

df.describe()
```

```python
Y = df["mpg"]
X = df[["weight", "year"]]

X = np.column_stack([np.ones(Y.shape[0]), X])
```

```python
# regression
b = np.linalg.pinv(X.T @ X) @ X.T @ Y

b
```

```python
# signifikant regression

k = len(b) - 1 # lÃ¤ngden pÃ¥ b minus 1, antalet parametrar

n = Y.shape[0] # antal observationer (stickprovsstorlek, lÃ¤ngden av y)

SSE = np.sum(np.square(Y - X @ b)) # Square sum of errors, vi multiplicerar vÃ¥ra fÃ¶rutsÃ¤gelser med

var  = SSE/(n-k-1) # variansen

S = np.sqrt(var) # uppskattning av standardavvikelsen 



Syy = (n*np.sum(np.square(Y)) - np.square(np.sum(Y)))/n # total varians fÃ¶r Y

SSR = Syy - SSE # vi tar den faktiska variansen och tar bort den frÃ¥n sum of square errors, detta ger oss SSE(Sum of square regression)

TSS = np.sum(np.square(Y - np.mean(Y))) # total sum of squares
Sxx = (n*np.sum(np.square(X)) - np.square(np.sum(X)))/n

# ensidigt test
sig_statistic = (SSR/k)/S # vi delar SSE med antalet parametrar och dividerar med S

sig_statistic # stort vÃ¤rde = signifikant, vi kan fÃ¶rkasta nollhypotesen


# "vad Ã¤r sannolikheten att vi ser detta vÃ¤rdet" # testar alla parametrar samtidigt
p_val = stats.f.sf(sig_statistic, k, n-k-1) # vi anvÃ¤nder oss av f-fÃ¶rdelningen fÃ¶r att rÃ¤kna ut p-vÃ¤rdet
p_val # sannolikheten att alla parametrar b skulle bara noll, i detta fall Ã¤r det vÃ¤ldigt liten
```
